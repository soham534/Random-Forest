# Splitting the data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

# Model Selection

```python
regressor = RandomForestRegressor()
```

# Hyper Parameter Tuning

```python
n_estimators = [int(i) for i in np.linspace(start=100,stop=1200,num=12)]
```

```python
max_features = ['auto','sqrt']
```

```python
max_depth = [int(i) for i in np.linspace(start=5,stop=30,num=6)]
```

```python
min_samples_split = [2,5,10,15,100]
```

```python
min_sample_leaf = [1,2,5,10]
```

```python
random_grid = {'n_estimators':n_estimators,
               'max_features':max_features,
               'max_depth':max_depth,
               'min_samples_split':min_samples_split,
               'min_samples_leaf':min_sample_leaf}
print(random_grid)
```

# Training the model

```python
# Parametric value
rf_regressoar = RandomizedSearchCV(estimator=regressor,param_distributions=random_grid,scoring='neg_mean_squared_error',n_iter=10,cv=5,verbose=2,random_state=42,n_jobs=1)
```

```python
rf_regressoar.fit(x_train,y_train)
```

```python
rf_regressoar.best_params_
```

# Testing the model

```python
y_pred = rf_regressoar.predict(x_test)
```

```python
plt.scatter(y_test,y_pred)
```

```python
finaldf = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
```

```python
finaldf
```

```python
sns.heatmap(finaldf.corr(),annot=True,cmap="Blues")
```

# Performance/Accuracy

```python
r2_score(y_test,y_pred)
```
